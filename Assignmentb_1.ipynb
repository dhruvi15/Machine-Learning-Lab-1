```python
# TODO 1: Load dataset
import pandas as pd

from google.colab import files
uploaded = files.upload()

# STEP 3: Read the CSV file into a DataFrame

df = pd.read_csv("assignment1_dataset.csv")
df

# Preview first 5 rows

df.head()

# Print dataset info, data types, and summary statistics

df.info()

df.describe()
```

```python
# TODO 2:

# 1. Check for missing values

Missing_count = df.isna().sum()
print("Missing Value :",Missing_count)

# 2. Fill missing WorkHours with the mean

# There no missing value in dataset

df["WorkHours"] = df["WorkHours"].fillna(df["WorkHours"].mean())

# 3. Fill missing EducationLevel with the mode

# There is no missing value in dataset

df["EducationLevel"] = df["EducationLevel"].fillna(df["EducationLevel"].mode()[0])

# 4. Remove duplicate rows if any

duplicates_count = df.duplicated().sum()
df_unique = df.drop_duplicates()

print("Duplicate  Value :",duplicates_count)
```

```python

# TODO 3:

# 1. Create WellnessIndex

df["WellnessIndex"] = df["WorkHours"] * df["SleepHours"]

# 2. Create AttendanceCategory

# High (>90)
# Medium (75â€“90)
# Low (<75)

def categorize_attendance(x):
    if x > 90:
        return "High"
    elif x >= 75 and x <= 90:
        return "Medium"
    else:
        return "Low"

df["AttendanceCategory"] = df["AttendanceRate"].apply(categorize_attendance)
```

```python
# TODO 4:
# Encode Gender and EducationLevel

from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt

le = LabelEncoder()

df["Gender_encoded"] = le.fit_transform(df["Gender"])                                       # Female -> 0   # MAle -> 1
df["EducationLevel_encoded"] = le.fit_transform(df["EducationLevel"])                       # Bachelor -> 1 # Master -> 2   # PhD -> 3

df.head()

# Convert RemoteWork (Yes/No) into numerical form

df["RemoteWork_encoded"] = le.fit_transform(df["RemoteWork"])                                # No -> 0       # Yes -> 1

# Scale WorkHours and AttendanceRate using StandardScaler or MinMaxScaler

scaler_StandardScaler = StandardScaler()

df[["WorkHours_scaled", "AttendanceRate_scaled"]] = scaler_StandardScaler.fit_transform(
    df[["WorkHours", "AttendanceRate"]])

# TODO 5:
# 1. Scatter plot: WorkHours vs ProductivityScore

plt.scatter(df["WorkHours"], df["ProductivityScore"])
plt.xlabel("WorkHours")
plt.ylabel("ProductivityScore")
plt.title("WorkHours vs ProductivityScore")
plt.show()

# 2. Boxplot: AttendanceCategory vs ProductivityScore

df.boxplot(column="ProductivityScore", by="AttendanceCategory")
plt.xlabel("AttendanceCategory")
plt.ylabel("ProductivityScore")
plt.title("AttendanceCategory vs ProductivityScore")
plt.suptitle("")
plt.show()

# 3. Correlation heatmap

corr = df.select_dtypes(include="number").corr()
plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()

# 4. Bar chart: EducationLevel vs ProductivityScore

edu_avg = df.groupby("EducationLevel")["ProductivityScore"].mean()
plt.figure()
plt.bar(edu_avg.index, edu_avg.values)
plt.xlabel("EducationLevel")
plt.ylabel("Average ProductivityScore")
plt.title("EducationLevel vs ProductivityScore")
plt.show()

# 5. Histogram of ProductivityScore

plt.hist(df["ProductivityScore"])
plt.xlabel("ProductivityScore")
plt.ylabel("Frequency")
plt.title("Distribution of ProductivityScore")
plt.show()

# 6. Scatter plot: WellnessIndex vs ProductivityScore

plt.scatter(df["WellnessIndex"], df["ProductivityScore"])
plt.xlabel("WellnessIndex")
plt.ylabel("ProductivityScore")
plt.title("WellnessIndex vs ProductivityScore")
plt.show()

# ___________SUMMARY_____________

# Productivity generally increases with higher work hours and attendance, showing that consistent effort plays an important role in performance.

# Employees with higher education levels tend to have better average productivity, which may be linked to experience and job responsibilities.

# The WellnessIndex has a positive but not perfectly linear relationship with productivity, indicating that balance helps, but other factors also matter.

```
